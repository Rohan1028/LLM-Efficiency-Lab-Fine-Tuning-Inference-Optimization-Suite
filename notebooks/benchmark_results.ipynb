# Benchmark Results
This notebook visualizes fine-tuning and inference metrics generated by the LLM Efficiency Lab.

```python
import json
from pathlib import Path
import pandas as pd
from plotly import express as px

quant_metrics = json.loads(Path('../results/inference/quantization_metrics.json').read_text())
df = pd.DataFrame(quant_metrics['results'])
px.bar(df, x='bits', y='throughput_tokens_per_s').show()
```
